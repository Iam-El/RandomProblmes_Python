_step1 : Enter Three strings . Three strings can be a combination of special characters , strings , numbers etc._

_An integer, greater than 1, that is only divisible by 1 and itself is called a prime number. All other numbers greater than 1 are called composite numbers. The integers 0 and 1 are neither prime nor composite. Write a python program that requests a positive integer from the user, determines if it is a prime, composite or neither prime or composite and prints the message. You can choose to use iterative loops to repeatedly run this script or have the user run the script for every input. Please specify instructions to run the code in markdown language above the code cell._


# _How to Run the program? :_

_Step1: Enter any positive integer._

_Step1:If you enter string or floating point number or negative number program handles it and prompts to enter a positive number again (Implemented a iterative Loop)_

```python
while True:
    try:
        #Enter any positive number
        input1=input("Enter positive Integer:")
        positiveInteger=int(input1) #typecast entered number to an integer
        if positiveInteger<0: #checks if entered number is negative
            print('Error: You did not enter a positive integer')
        else:
            if positiveInteger==0 or positiveInteger==1: #checks if entered number is 0 or 1
                print('It is neither prime nor composite')
            elif positiveInteger>1: #checks if entered integer is greater than 1 
                for i in range(2,positiveInteger):
                    if positiveInteger%i==0: #checks if the number is prime or composite
                        print('It is a composite')
                        break
                else:
                    print('It is a prime')
                
    except ValueError:
        print('Error: Entered number is not an integer!!')

```

# _How to Run the program? :_

_Step1: Enter x1 , y1, x2, y2 (Cordinates of rectangle) .Enter x ,y (test corniates). If you enter string program handles it and re run the program._

_Step2: After you enter the cordinates it checks if cordinate lie within a rectangle . and return True or false._

```python
def test_rectangle(x1, y1, x2, y2, x, y):  #Function to check if the given cordinates fall within a given rectangle

    if ((x1 <= x and x <=x2) or (x2 <= x and x <=x1)) and ((y1 <= y and y <=y2) or (y2 <= y and y <=y1)):
        print(True)
    else:
        print(False)


finalInput = "Y"
while finalInput == 'N' or finalInput == 'Y':
    try:
        #input the cordinates x1,x2,y1,y2,x and y
        
        input1 = input('Enter x1:')
        x1_cordinate = float(input1)
        input2 = input('Enter y1:')
        y1_cordinate = float(input2)
        input3 = input('Enter x2:')
        x2_cordinate = float(input3)
        input4 = input('Enter y2:')
        y2_cordinate = float(input4)
        input5 = input('Enter x:')
        x_cordinate = float(input5)
        input6 = input('Enter y:')
        y_cordinate = float(input6)


        # function call test_rectangle 
        test_rectangle(x1_cordinate, y1_cordinate, x2_cordinate, y2_cordinate, x_cordinate, y_cordinate)

        #Prompts the user to test the inputs
        finalInput = input('Do you want to continue(Y/y or N/n):')
        #if input N then says goodbye and comes out of the loop
        if finalInput == 'N':
            print('Goodbye!')
            break
        #if input is Y it prompts the user to input the values
        elif finalInput == 'Y':
            print('continue')
        #if the input is other than Y or N then exit the program
        else:
            print("Type Y to continue or N to exit with goodbye!!")
            exit
    except ValueError:
        print('Error:Enter Integer or Floating point number')



```
