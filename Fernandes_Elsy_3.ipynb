_(25 points) A string is an anagram of another if the second string is simply a scrambled version of the first. Write a python program to implement the following game:_

_a) Reads in a file that has words and their meanings in a text file. An example “words and their meanings” file is given in blackboard. Note that your program needs to ask the user for the “words and their meanings” file to use._

_b) The words and their meanings text file is of the csv (comma separated values) format. Use either notepad++ or notepad to create your file in the same format as the mywords.txt file given to you in blackboard_

_c) Your program should then pick a word from the “words and their meanings” file, scramble the letters, and ask the user to unscramble it. Every run of your program should pick a word at random._

_d) The user may type in the unscrambled word or may ask for the definition/meaning of the word by entering a question mark. The game continues until the user says “no” to the question: “Do you want to continue?”. A sample run is shown below (user inputs are in red):_

# _Steps_#

_1. When program is run give a filename mywords.txt. Program handles if you provide the file name it doesnt exist. Rerun the program_

_2. mywords.txt has a format of one key and a value format. If a file with any other format is given Program handles it . ( For example change the content of file as aaaaaa,ddddddd,hhjhdjaj ). rerun the program_

_3. Enter any empty filename . Program handles it . Rerun the program._

_4. Enter a correct filename with correct data . Program will prompt for the answer. If you type ? it will give the meaning for the word._

_5. If you enter a wrong answer , it says you have entered a wrong answer and ask you to try again_

_6. If you enter a correct answer it will prompt you 'Do you want to cntinue or not?' you can either type yes or no(as mentioned in the pdf) . If you type other than yes or no . it will prompt you again 'Do you want to continue or not?'_

```python
from random import shuffle
import re
import os
import sys
import random

c = 0
file_name = input('Give the name of the “words and their meanings” file:')  # give the filename
pattern = r"(.+?)(?:,|$)"
incorrect_file_formatting = False
try:
    file = open(file_name)  # open file in read mode
    d = {} #Initialize empty dictionary
    for line in file:
        words = line.replace('\n', '').split(",") # split the word by comma
        key = words[0] # seperate out key
        value = words[1] #seperate out value
        d[key] = value # put them as a key value pair in a dictionary
        if len(words)!=2: #checks the length of dictionary is not equal to 2 then its not a properly formatted file
            incorrect_file_formatting = True
    if len(d) == 0: # if length of dictionary is zero then the file is empty
        print("File is empty")
    elif incorrect_file_formatting:
        print("File is not properly formatted") # prints a file is not properly formatted
    else:
        inputvalue = 'yes'
        while inputvalue == 'yes':
            count = 0
            rw = random.choice(list(d.keys())) # generate the random word from a dictionary keys
            word = list(rw) # assign it to the variable
            shuffle(word) # use python shuffle library scrumble the key
            sw = ''.join(word) # join the word
            print("Unscramble the following letters to form a word. Type “?” for the meaning of the unscrambled word:")
            print(sw) # print the scrumbled word
            for i in range(len(rw)):
                a = input("Enter the answer [or ? for the meaning]:")
                if a == '?':
                    count = count + 1
                    print("The word means:", d.get(rw)) #get the meaning of the word when user enters ?
                elif d.get(a) == d.get(rw): # if the user entered key matching the crumbled word
                    inputvalue = input("You got it! Do you want to continue [yes or no]:")
                    while inputvalue != "yes" and inputvalue != "no": #providing the option for the user to continue or not
                        inputvalue = input("You got it! Do you want to continue [yes or no]:")
                    if inputvalue == 'no': #if provided input is no then exits the program
                        print('Goodbye!!')
                    break
                else:
                    print("Wrong, Try Again!")
except:
    e = sys.exc_info()[0]
    print('File does not exist!! or incorect file format') #if file does not exist or file is not in a correct format
```
